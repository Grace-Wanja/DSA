- 2. **Abstract Data Types (ADT) and Data Structures**
- An ADT is a model that defines operations and behaviors for a data type without specifying how these are implemented or how data is stored. ADT only mentions what operations are to be performed but not how these operations will be implemented. It does not specify how data will be organized in memory and what algorithms will be used for implementing the operations. It is called “abstract” because it provides an implementation-independent view. They are known as adstract because they only provide the essentials and hide the details a process called abstruction.

- i). The Role of Abstract Data Types in Programming
  - ADTs are used to create fast algorithms in less time.
  - they help to manage and organize the data.
  - they make code clean and easy to understand.
  - they make the execution program fast and consume less time
- A user does not need to know the hidden implementation as long as the data is being entered and obtained correctly.

- ii). Difference between ADTs and Data Structures
  - An Abstract Data Type is a conceptual(basic layout) model that defines the operations and behavior of a data type without specifying how it's implemented, while
 a Data structure is the concrete implementation of that data type, detailing how the data is stored and accessed in memory. 
  - An ADT describes what a data structure can do, while the data structure defines how it does it. 
  - An ADT defines the operations that can be performed on data (like adding, removing, searching) without specifying the underlying mechanics while a data structure focuses on implementation such as how to organize data in memory, like using an array, linked list, or tree, to efficiently execute the operations defined by an ADT. 

- iii). Examples of ADTs and how they are implemented in programming languages
   - a)List ADT
    - The List ADT needS to store the required data in a sequence and should have the following operations:
      - get():to return an element from the list at any given position.
      - insert():to insert an element at any position in the list.
      - remove():to remove the first occurrence of any element from a non-empty list.
      - removeAt():to remove the element at a specified location from a non-empty list.
      - replace():to replace an element at any position with another element.
      - size():to return the number of elements in the list.
      - isEmpty():to return true if the list is empty; otherwise, return false.
      - isFull():to return true if the list is full; otherwise, return false

   - b)Stack ADT
    - In Stack ADT, the order of insertion and deletion should be according to the FILO or LIFO Principle. Elements are inserted and removed from the same end, called the top of the stack. It should also support the following operations:
      - push(): Insert an element at one end of the stack called the top.
      - pop(): Remove and return the element at the top of the stack, if it is not empty.
      - peek(): Return the element at the top of the stack without removing it, if the stack is not empty.
      - size(): Return the number of elements in the stack.
      - isEmpty(): Return true if the stack is empty; otherwise, return false.
      - isFull(): Return true if the stack is full; otherwise, return false.

   - c)Queue ADT
     - The Queue ADT follows a design similar to the Stack ADT, but the order of insertion and deletion changes to FIFO. Elements are inserted at one end (called the rear) and removed from the other end (called the front). It should support the following operations:
      - enqueue(): Insert an element at the end of the queue.
      - dequeue(): Remove and return the first element of the queue, if the queue is not empty.
      - peek(): Return the element of the queue without removing it, if the queue is not empty.
      - size(): Return the number of elements in the queue.
      - isEmpty(): Return true if the queue is empty; otherwise, return false


3. **Types of Data Structures**
- Arrays; An array is a data structure used to store multiple elements. Arrays are indexed, meaning that each element in the array has an index, a number that says where in the array the element is located with first element being at index 0
- Linked Lists; A Linked List is, a list where the nodes are linked together. Each node contains data and a pointer. The way they are linked together is that each node points to where in the memory the next node is placed. In a Linked List the;
  node is a structure that contains the following:
   - data(value stored in the node)
   - pointer(reference to the next node)

   - head-the first node in a linked list
   - tail-the last node in a linked list
- Stacks; A stack is a data structure that can hold many elements
  - Stacks can be implemented by using arrays or linked lists. uses the LIFO(last in first out) style of organizing data
- Queues; A queue is a data structure that can hold many elements.
  - Queues can be implemented by using arrays or linked lists. Queues use the FIFO(First In First Out) way of organizing elements
- Recursion; Recursion is a programming technique where a function calls itself repeatedly until it reaches a base case that stops the recursion
- Trees; A tree is a hierarchical data structure, where each node has a value and zero or more child nodes. a child node is a node that is directly connected to another node, called the parent node. The Tree data structure is similar to Linked Lists in that each node contains data and can be linked to other nodes.
- Graphs; A Graph is a non-linear data structure that consists of vertices (nodes) and edges.A vertex, also called a node, is a point or an object in the Graph, and an edge is used to connect two vertices with each other. Graphs are non-linear because the data structure allows us to have different paths to get from one vertex to another, unlike with linear data structures like Arrays or Linked Lists.

- 4. **Operations on Data Structure**
    Define and explain the following operations in data structures
    -  Traversing: This operation involves visiting each element in a data structure once, typically to perform some action on each element like printing or calculating a value based on it. 
    - Searching: This operation aims to find a specific element within the data structure based on a given key or value, returning its location if found or indicating its absence. 
    - Insertion: Adding a new element to a data structure at a designated position, potentially requiring adjustments to maintain the structure's integrity. 
    - Deletion: Removing an existing element from the data structure, which may involve shifting elements to fill the gap left behind. 
    - Updating: Modifying the value of an existing element within the data structure. 
    - Sorting: Arranging the elements of a data structure in a specific order, such as ascending or descending, based on a chosen comparison criteria. 
    - Merging: Combining two already sorted lists into a single sorted list, preserving the order of elements from both list

```python
##Examples of how the operations are implemented
#Traversing
arr = [10, 20, 30, 40]
for elem in arr:
    print(elem)

#Searching
arr = [10, 20, 30, 40]
target = 30
for index, value in enumerate(arr):
    if value == target:
        print(f"Found at index {index}")
        break

#Insertion
arr = [10, 20, 30, 40]
arr.append(50)  # Inserts 50 at the end
print(arr)  # Output: [10, 20, 30, 40, 50]

#Deletion
arr = [10, 20, 30, 40]
arr.remove(30)  # Removes element 30
print(arr)  # Output: [10, 20, 40]

#Updating
arr = [10, 20, 30, 40]
arr[2] = 35  # Updates element at index 2
print(arr)  # Output: [10, 20, 35, 40]

#Sorting
arr = [30, 10, 40, 20]
arr.sort()  # Sorts the array in ascending order
print(arr)  # Output: [10, 20, 30, 40]

#Merging 2 arrays
arr1 = [10, 20, 30]
arr2 = [15, 25, 35]
merged = arr1 + arr2  # Merging two lists
merged.sort()  # Sorting the merged list
print(merged)  # Output: [10, 15, 20, 25, 30, 35]


```

- 5. **Algorithms**
An algorithm is a set of step-by-step instructions to solve a given problem or achieve a specific goal.
Algorithms are fundamental to computer programming as they provide step-by-step instructions for executing tasks. An efficient algorithm can help us to find the solution we are looking for, and to transform a slow program into a faster one.

- A pseudocode for the summation of numbers in an array algorithm
- FUNCTION sumArray(arr):
    - sum = 0
    - FOR i FROM 0 TO length(arr) - 1:
        - sum = sum + arr[i]
    - RETURN sum

- This pseudocode defines a function matrixMultiply that takes two matrices A and B as input and returns their product. The function first checks if the number of columns in A is equal to the number of rows in B, which is a necessary condition for matrix multiplication. If the matrices are incompatible, the function throws an error.

- FUNCTION matrixMultiply(A, B):
    - IF number of columns in A != number of rows in B:
        - THROW ERROR "Incompatible matrices for multiplication"
    
    - result = NEW MATRIX with number of rows in A and number of columns in B
    - FOR i FROM 0 TO number of rows in A - 1:
        - FOR j FROM 0 TO number of columns in B - 1:
            - FOR k FROM 0 TO number of columns in A - 1:
                - result[i, j] = result[i, j] + A[i, k] * B[k, j]
    - RETURN result



- Brute Force Algorithm;
This is the most basic and simplest type of algorithm. A Brute Force Algorithm is the straightforward approach to a problem i.e., the first approach that comes to our mind on seeing the problem. More technically it is just like iterating every possibility available to solve that problem

- Recursive Algorithm
This type of algorithm is based on recursion. In recursion, a problem is solved by breaking it into subproblems of the same type and calling own self again and again until the problem is solved with the help of a base condition. 
   - forexample finding the factorial of a number
- Divide and Conquer Algorithm 
In Divide and Conquer algorithms, the idea is to solve the problem in two sections, the first section divides the problem into subproblems of the same type. The second section is to solve the smaller problem independently and then add the combined result to produce the final answer to the problem
   - used in merge sort quick sort binary search 
- Dynamic Programming Algorithm 
This type of algorithm is also known as the memoization technique because in this the idea is to store the previously calculated result to avoid calculating it again and again. In Dynamic Programming, divide the complex problem into smaller overlapping subproblems and store the result for future use.
   - used to solve Knapsack Problem, Weighted Job Scheduling
- Greedy Algorithm 
In the Greedy Algorithm, the solution is built part by part. The decision to choose the next part is done on the basis that it gives an immediate benefit. It never considers the choices that had been taken previously.
   - used for Dijkstra Shortest Path Algorithm, Prim’s Algorithm, Kruskal’s Algorithm, Huffman Coding, etc
- Backtracking Algorithm 
In Backtracking Algorithm, the problem is solved in an incremental way i.e. it is an algorithmic technique for solving problems recursively by trying to build a solution incrementally, one piece at a time, removing those solutions that fail to satisfy the constraints of the problem at any point of time.
  - used for the Hamiltonian Cycle, M-Coloring Problem, N Queen Problem, Rat in Maze Problem, etc.

- 7. Guidelines for Flowcharts
Summarize the guidelines for drawing flowcharts, including standard 
symbols and best practices. 
- **Guidlines and rules used when drawing flowcharts**
- Rule 1: Flowchart opening statement must be ‘start’ keyword.
- Rule 2:  Flowchart ending statement must be ‘end’ keyword.
- Rule 3: All symbols in the flowchart must be connected with an arrow line.
- Rule 4: Each decision point should have two or more distinct outcomes.
- Rule 5: Flow should generally move from top to bottom or left to right
- *Standard symbols for drawing flowcharts*
   - Oval or Rounded Rectangle: Represents the start or end of a process.
   - Rectangle: Represents a process, action, or operation step.
   - Diamond: Represents a decision or branching point.
   - Parallelogram: Represents input or output.
   - Cylinder: Represents a database or stored data.
   - Rectangle with Double Sides: Represents a predefined process or subroutine.
   - Circle: Represents a connector or continuation point.
- Best Practices for Drawing Flowcharts: 
- - Start with the Start symbol and finish with the End symbol.
- - use standard symbols for each operation.
- - Follow a logical flow from one step to the next.
- - Label decision points clearly (e.g., Yes/No branches).
- - Avoid crossing arrows to maintain clarity.
- - Keep it simple: Break down complex tasks into smaller flowcharts if needed.
- - Ensure readability: Consistent spacing and clear labels are key.
- - Review the flowchart for correctness and completeness.
