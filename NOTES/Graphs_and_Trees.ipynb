## 1. Introduction to Graphs (10 Marks)

#### Graphs Definition:
A Graph is a non-linear data structure consisting of a finite set of vertices(or nodes) and a set of edges(or links)that connect a pair of nodes. Graphs are widely used to represent relationships between entities. Graph is composed of a set of vertices( V ) and a set of edges( E ). The graph is denoted by G(V, E).
- V represents the set of vertices.
- E represents the set of edges connecting vertices.
    - NOTE: trees are a subset of the graphs and also lists are special cases of trees and thus of graphs, i.e. the graphs represent a generalized structure that allows modeling of very large set of real-world situations.

- Applications: Social networks, maps, recommendation systems

#### Types of Graphs:
1. **Directed vs Undirected**: 
    - Directed graphs have edges with a direction . A graph in which edges have a direction, i.e., the edges have arrows indicating the direction of traversal (e.g., Twitter followers).
    - Undirected graphs have edges with no direction . An undirected graph is a graph where edges do not have a specific direction, meaning connections go both ways. If two places are connected, you can travel in either direction. Examples include friendships on social media and two-way roads (e.g., Facebook friendships).

2. **Weighted vs Unweighted**: 
    - A weighted graph is a graph where each edge has a number (weight) that represents distance, cost, or time. These graphs help find the shortest or cheapest paths. Examples include Google Maps, airline routes, and delivery networks ie, Weighted graphs assign weights (costs) to edges.
    - An unweighted graph is a graph where all edges are treated equally, with no extra values like distance or cost. It simply shows connections between points. Examples include basic social networks and metro maps without travel times ie, Unweighted graphs do not assign weights.

3. **Connected vs Disconnected**: 
    - A connected graph has a path between any pair of vertices.
    - A disconnected graph has at least two vertices with no path between them.
    - ie, Graph G is said to be connected if any pair of vertices (Vi, Vj) of a graph G is reachable from one another. Or a graph is said to be connected if there exists at least one path between each and every pair of vertices in graph G, otherwise, it is disconnected. A null graph with n vertices is a disconnected graph consisting of n components. Each component consists of one vertex and no edge. 

4. **Cyclic vs Acyclic**: 
    - A cyclic graph contains at least one cycle.
    - An acyclic graph has no cycles.

5. **Special Graphs**: 
    - **Bipartite**: A graph where the vertices can be divided into two disjoint sets.
    - **Planar**: A graph that can be drawn on a plane without edges crossing.


## Graphs vs Trees: When to Use Each Structure
### Comparison Table

| **Feature**  | **Graph**  | **Tree**  |
|-------------|-----------|---------|
| **Definition** | A collection of nodes (vertices) connected by edges. | A specialized graph where nodes are hierarchically arranged, with one root and parent-child relationships. |
| **Structure** | Can be cyclic or acyclic, directed or undirected. | Always acyclic (except for general trees like N-ary trees) and hierarchical. |
| **Connectivity** | Any node can connect to any other node. | Nodes have a single parent except for the root. |
| **Types** | Directed Graphs, Undirected Graphs, Weighted Graphs, DAGs (Directed Acyclic Graphs), etc. | Binary Trees, Binary Search Trees (BST), AVL Trees, B-Trees, etc. |
| **Efficiency** | Can model complex relationships but may require extra processing to maintain constraints. | Efficient for hierarchical data but has limited connectivity. |
| **Storage Complexity** | Adjacency list (O(V + E)) or adjacency matrix (O(V²)). | Usually O(n) since there are n-1 edges in a tree with n nodes. |
| **Search Algorithms** | BFS, DFS, Dijkstra’s, A* | Preorder, Inorder, Postorder, Level-order Traversal |
| **Common Uses** | Networks (social, transportation, computer networks), recommendation systems, dependency resolution, AI pathfinding. | Hierarchical data (filesystems, databases, organizational structures), search trees, decision trees. |

### When to Use a Graph
- **When relationships between elements are complex and non-hierarchical** (e.g., social networks, web pages with hyperlinks).
- **When you need to represent connections with varying constraints** (e.g., shortest path in a map, circuit design).
- **When cycles are needed** (e.g., modeling game states, state machines).
- **For AI and machine learning applications** (e.g., knowledge graphs, neural networks).

### When to Use a Tree
- **When data has a natural hierarchical structure** (e.g., folder structure in a file system, organization charts).
- **When fast search, insert, and delete operations are needed** (e.g., binary search trees, AVL trees).
- **For decision-making processes** (e.g., decision trees in AI, expression trees in compilers).
- **For efficient sorting and retrieval** (e.g., B-Trees in databases, heaps in priority queues).

## 4. Graph Traversal Algorithms (10 Marks)

#### Graph Traversal:
1. **Breadth-First Search (BFS)**
- BFS explores the graph level by level, starting from a given source vertex. It uses a queue to keep track of vertices to visit next. 
- steps
   - Start at the source vertex and mark it as visited.
   - Add the source vertex to a queue.
   - While the queue is not empty:
   - Dequeue a vertex from the queue.
   - Visit all its unvisited neighbors, mark them as visited, and enqueue them.

- applications 
    - Shortest path in unweighted graphs.
    - Social network analysis (e.g., finding friends within a certain distance).

2. **Depth-First Search (DFS)**: Explores the graph deeply before backtracking.
- DFS explores the graph as deep as possible before backtracking. It uses a stack (recursion or explicit stack) to keep track of vertices to visit next.

- steps
     - Start at the source vertex and mark it as visited.
     - Recursively visit all unvisited neighbors of the current vertex.

- Applications:
   - Cycle detection in graphs.
   - Topological sorting.
   - Solving puzzles (e.g., mazes).


```python
from collections import deque

# BFS Implementation
def bfs(graph, start):
    visited = set()  # Set to keep track of visited vertices
    queue = deque([start])  # Queue for BFS traversal
    traversal_order = []  # List to store the order of traversal

    while queue:
        vertex = queue.popleft()  # Dequeue a vertex
        if vertex not in visited:
            visited.add(vertex)
            traversal_order.append(vertex)
            # Enqueue all unvisited neighbors
            queue.extend(neighbor for neighbor in graph[vertex] if neighbor not in visited)
    
    return traversal_order

# DFS Implementation
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()  # Set to keep track of visited vertices
    traversal_order = []  # List to store the order of traversal

    if start not in visited:
        visited.add(start)
        traversal_order.append(start)
        # Recursively visit all unvisited neighbors
        for neighbor in graph[start]:
            traversal_order.extend(dfs(graph, neighbor, visited))
    
    return traversal_order

# Example usage
graph = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 5],
    3: [1],
    4: [1, 5],
    5: [2, 4]
}

print("BFS Traversal Order:", bfs(graph, 0))
print("DFS Traversal Order:", dfs(graph, 0))
```

## 5. Shortest Path Algorithms (10 Marks)

#### Shortest Path Algorithms:
1. **Dijkstra's Algorithm**:
    - Finds the shortest path between nodes in a graph with non-negative edge weights.
#### Dijkstra’s Algorithm (Implementation Required)

Dijkstra's Algorithm is used to find the shortest path from a source node to all other nodes in a weighted graph with non-negative edge weights. The algorithm works by iteratively selecting the node with the smallest tentative distance, marking it as visited, and updating the distances to its neighbors.

**Steps of Dijkstra’s Algorithm:**
1. Initialize the distance to the source node as 0 and the distance to all other nodes as infinity.
2. Mark all nodes as unvisited.
3. Select the unvisited node with the smallest tentative distance.
4. For each neighbor of the current node, calculate the tentative distance from the source and update it if it’s smaller than the current known distance.
5. Mark the current node as visited and repeat the process until all nodes have been visited.

**Time Complexity:**
- **O(V²)** for a simple implementation.
- Can be optimized to **O((V + E) log V)** using a priority queue.

```python

import heapq

def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    pq = [(0, start)]
    
    while pq:
        current_distance, current_node = heapq.heappop(pq)
        
        if current_distance > distances[current_node]:
            continue
        
        for neighbor, weight in graph[current_node]:
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))
    
    return distances

# Example usage
graph = {
    1: [(2, 1), (3, 4)],
    2: [(1, 1), (3, 2)],
    3: [(1, 4), (2, 2)]
}
print(dijkstra(graph, 1))
```

## 6. Minimum Spanning Trees (MST) (5 Marks)

#### Minimum Spanning Tree (MST):
An MST is a subset of edges that connects all vertices with the minimum possible total edge weight and no cycles.
#### Definition of Minimum Spanning Tree (MST)

A **Minimum Spanning Tree (MST)** of a connected, undirected graph is a tree that spans all the vertices of the graph and has the minimum possible total edge weight. In other words, it is a subset of the edges of the graph that connects all the vertices together, without forming any cycles, and with the minimum total edge weight.

### Kruskal’s Algorithm

Kruskal’s algorithm is a greedy algorithm used to find the Minimum Spanning Tree of a graph. It works by sorting all the edges in the graph by their weight in increasing order and then adding the edges one by one to the MST, provided they don’t form a cycle.

**Steps of Kruskal's Algorithm:**
1. Sort all the edges in the graph in increasing order of their weights.
2. Initialize a disjoint-set data structure to keep track of connected components.
3. For each edge in sorted order:
   - If the edge connects two different components, add it to the MST.
   - Otherwise, discard it (to avoid cycles).
4. Repeat until the MST contains **V - 1** edges (where **V** is the number of vertices).

**Time Complexity:**
- Sorting edges takes **O(E log E)**, where **E** is the number of edges.
- The union-find operations take **O(E log V)** (using path compression and union by rank).
- Overall complexity: **O(E log E)** or **O(E log V)**.



```python
#### Kruskal’s Algorithm:
def kruskal(graph, vertices):
    edges = [(weight, u, v) for u in graph for v, weight in graph[u]]
    edges.sort()  # Sort edges by weight
    
    parent = {v: v for v in vertices}
    rank = {v: 0 for v in vertices}
    
    def find(v):
        if parent[v] != v:
            parent[v] = find(parent[v])
        return parent[v]
    
    def union(u, v):
        root_u = find(u)
        root_v = find(v)
        if root_u != root_v:
            if rank[root_u] > rank[root_v]:
                parent[root_v] = root_u
            else:
                parent[root_u] = root_v
                if rank[root_u] == rank[root_v]:
                    rank[root_v] += 1
    
    mst = []
    for weight, u, v in edges:
        if find(u) != find(v):
            mst.append((u, v, weight))
            union(u, v)

            return mst

# Example usage
graph = {
    1: [(2, 1), (3, 4)],
    2: [(1, 1), (3, 2)],
    3: [(1, 4), (2, 2)]
}
vertices = [1, 2, 3]
print(kruskal(graph, vertices))
```

### 8. Binary Search Trees (BST) (10 Marks)

#### Definition of Binary Search Tree (BST)

A **Binary Search Tree (BST)** is a binary tree in which each node has at most two children, and for every node:
- The left subtree contains only nodes with values less than the node’s value.
- The right subtree contains only nodes with values greater than the node’s value.
- Both the left and right subtrees must also be binary search trees.

#### Properties of Binary Search Tree

- Left Subtree: All values are less than the root.
- Right Subtree: All values are greater than the root.
- No Duplicate Nodes: Typically, BSTs do not allow duplicate values.
- Inorder Traversal: Produces a sorted sequence of values.

- **Unique values**: Each node contains a unique value, which is used to organize the tree.
- **Ordered structure**: For any node in the tree:
  - The values in the left subtree are less than the node’s value.
  - The values in the right subtree are greater than the node’s value.
- **Efficient searching**: The BST allows for efficient searching, insertion, and deletion operations in O(log n) time in the average case.




```python
#Python implementations of insertion, deletion, and search operations for a BST
def insert(root, key):
    #Insert a key into the BST
    if root is None:
        return {"key": key, "left": None, "right": None}
    if key < root["key"]:
        root["left"] = insert(root["left"], key)
    elif key > root["key"]:
        root["right"] = insert(root["right"], key)
    return root

def search(root, key):
    #Search for a key in the BST.
    if root is None or root["key"] == key:
        return root
    if key < root["key"]:
        return search(root["left"], key)
    return search(root["right"], key)

def search(root, key):
    #Search for a key in the BST.
    if root is None or root["key"] == key:
        return root
    if key < root["key"]:
        return search(root["left"], key)
    return search(root["right"], key)

def delete(root, key):
    #Delete a key from the BST.
    if root is None:
        return root
    if key < root["key"]:
        root["left"] = delete(root["left"], key)
    elif key > root["key"]:
        root["right"] = delete(root["right"], key)
    else:
        # Node with only one child or no child
        if root["left"] is None:
            return root["right"]
        elif root["right"] is None:
            return root["left"]
        # Node with two children: Get the inorder successor (smallest in the right subtree)
        root["key"] = min_value(root["right"])
        # Delete the inorder successor
        root["right"] = delete(root["right"], root["key"])
    return root

def min_value(node):
    #Find the minimum value in a subtree.
    while node["left"] is not None:
        node = node["left"]
    return node["key"]

def inorder_traversal(root):
    #Perform inorder traversal of the BST.
    result = []
    if root:
        result = inorder_traversal(root["left"])
        result.append(root["key"])
        result += inorder_traversal(root["right"])
    return result


# Create a BST
root = None
root = insert(root, 50)
root = insert(root, 30)
root = insert(root, 20)
root = insert(root, 40)
root = insert(root, 70)
root = insert(root, 60)
root = insert(root, 80)

# Inorder Traversal
print("Inorder Traversal:", inorder_traversal(root))

# Search for a key
print("Search for 40:", "Found" if search(root, 40) else "Not Found")
print("Search for 100:", "Found" if search(root, 100) else "Not Found")

# Delete a key
root = delete(root, 20)
print("Inorder Traversal after deleting 20:", inorder_traversal(root))

root = delete(root, 30)
print("Inorder Traversal after deleting 30:", inorder_traversal(root))

root = delete(root, 50)
print("Inorder Traversal after deleting 50:", inorder_traversal(root))
```

## 9. Balanced Trees & Heaps (10 Marks)

#### Balanced Trees:
- **AVL Trees**: These trees automatically maintain balance by ensuring that the heights of left and right subtrees differ by at most 1.
- **Red-Black Trees**: A balanced binary search tree where each node has a color (red or black) that maintains the balance of the tree.

#### Heaps:
- **Heap**: A binary tree with the property that the key of each node is either greater (in a max-heap) or smaller (in a min-heap) than the keys of its children.
- **Priority Queue**: A data structure that supports efficient insertion and extraction of the element with the highest (or lowest) priority.



```python
##### Priority Queue using a Min-Heap:

import heapq

# Min-Heap Priority Queue
pq = []
heapq.heappush(pq, (1, "task1"))
heapq.heappush(pq, (3, "task3"))
heapq.heappush(pq, (2, "task2"))

print("Priority Queue (Min-Heap):")
while pq:
    priority, task = heapq.heappop(pq)
    print(f"{task} with priority {priority}")
```

```python
# 9. Balanced Trees & Heaps
def heapify(arr):
    heapq.heapify(arr)
    return arr
heap = [3, 1, 4, 1, 5]
print("Heap:", heapify(heap))
```

#### B-Trees:

- Used in databases and file systems.

#### Tries (Prefix Trees):
- A tree used to store a dynamic set of strings where the keys are usually strings.
- Useful in tasks like autocomplete or spell-checking.

#### N-ary Trees:
- A tree in which each node can have at most N children.

#### Segment Trees:
- A tree used for storing intervals or segments, and it allows querying which segments overlap with a given point or interval.

```python
##### Trie (Prefix Tree) Implementation:
# 10. Tries (Prefix Trees)
def insert_trie(trie, word):
    node = trie
    for char in word:
        if char not in node:
            node[char] = {}
        node = node[char]
    node['*'] = True

def search_trie(trie, word):
    node = trie
    for char in word:
        if char not in node:
            return False
        node = node[char]
    return '*' in node

trie = {}
insert_trie(trie, "hello")
print("Search 'hello':", search_trie(trie, "hello"))
print("Search 'hell':", search_trie(trie, "hell"))
```

## 11. Complexity & Real-World Applications (10 Marks)

#### Time Complexity of Tree Operations:
1. **Insertion** and **Search**: O(log n) in balanced trees, O(n) in unbalanced trees.
2. **Deletion**: O(log n) for balanced trees, O(n) for unbalanced trees.
3. **Traversal**: O(n), as every node is visited once.

#### Importance of Balanced Trees:
- Balanced trees like AVL and Red-Black trees maintain efficient operations even with large datasets, ensuring O(log n) time complexity for insertions, deletions, and lookups.

#### Real-World Applications of Trees:
1. **File System Navigation**: Trees are used to represent the hierarchical structure of files and directories.
2. **Autocompletion**: Tries are widely used in text prediction and autocompletion applications.
